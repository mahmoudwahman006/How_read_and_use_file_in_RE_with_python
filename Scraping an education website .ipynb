```python
import requests
from bs4 import BeautifulSoup
coun=0
list_headlink=[]
list_summary=[]
list_headline=[]
coun_line=0 #counter of line
coun_link=0 # counter of link
coun_s=0 # counter of summary
for i in range(1,18):

    website=requests.get(f"https://coreyms.com/page/{i}").content
    soup =BeautifulSoup(website,'html.parser')

    # scrap the link of the video
    articles=soup.find_all('iframe',{'class':'youtube-player'})
    for article in articles :
        headlink = article.attrs['src']
        #print(headlink)
        list_headlink.append(headlink)
        coun_link += 1

    # scrap the line of the packet (the head line)
    articles=soup.find_all('a',{'class':'entry-title-link'})
    for article in articles :
        headline = article.text
        list_headline.append(headline)
        coun_line += 1

    # scrap the link of the summary
    articles=soup.find_all('div',{'class':'entry-content'})
    for article in articles :
        summary = article.p.text
        list_summary.append(summary)
        coun_s+=1

    # counter -1 becuase of the list is out of range
    #visualize the stored data 
    for x in range(coun_link-1):
        print(list_headline[x])
        print(" ")
        print(list_summary[x])
        print(" ")
        print(list_headlink[x])
        print(" ")
        print(" ")
        print(" ")


'''print("summary= ",coun_s)
print("line= ",coun_line)
print("link ", coun_link)'''


## Ladies and gentlemen it was an honer to pleasure U ^__^  
```
